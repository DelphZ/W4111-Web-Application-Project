**Run all tests as a whole**: The notebook just demonstrates all tests I have written in `test_classic_model_end_points.py`, so instead of running all cells of notebooks, we can also run all tests in a single run by using `python -m pytest`.

# A full lifetime of a customer
## Post(Create) -> Get -> Update -> Delete of a customer
Involve testing /customers/{customerNumber}	for GET, PUT, DELETE and /customers for POST

````python
# create
payload = {
    "customerNumber": 2021,
    "customerName": "Unit Test Corp",
    "contactLastName": "Tester",
    "contactFirstName": "Joe",
    "phone": "555-0199",
    "addressLine1": "123 Code Lane",
    "city": "Testville",
    "country": "USA"
}
create_res = requests.post(f"{BASE_URL}/customers", json=payload)
assert create_res.status_code, 200
new_id = create_res.json()
print(f"   Created Customer ID: {new_id}")
print(f"   Created Customer Name: {payload["customerName"]}")

# get
get_res = requests.get(f"{BASE_URL}/customers/{new_id}")
assert get_res.status_code, 200
assert get_res.json()["customerName"], "Unit Test Corp"
assert get_res.json()["city"], "Testville"

# update
update_payload = payload.copy()
update_payload["customerName"] = "Updated Corp"
put_res = requests.put(f"{BASE_URL}/customers/{new_id}", json=update_payload)
assert put_res.status_code, 200

# verify get
get_res = requests.get(f"{BASE_URL}/customers/{new_id}")
assert get_res.status_code, 200
assert get_res.json()["customerName"], "Updated Corp"
print(f"   Updated Customer Name: {get_res.json()["customerName"]}")

# D. DELETE
del_res = requests.delete(f"{BASE_URL}/customers/{new_id}")
assert del_res.status_code, 200

# E. VERIFY DELETE (Error Case)
check_res = requests.get(f"{BASE_URL}/customers/{new_id}")
assert check_res.status_code, 404
print(f"   Get return 404 after delete the created customer successfully. Verification done!")
````

# A full lifetime of orders
## Create -> Get -> Update -> Delete of an order
Involve testing /orders/{orderNumber} for GET, PUT, DELETE and /orders for POST.

````python
payload = {
    "orderDate": "2025-01-01",
    "requiredDate": "2025-01-10",
    "status": "In Process",
    "customerNumber": EXISTING_CUSTOMER_ID
}
create_res = requests.post(f"{BASE_URL}/orders", json=payload)
assert create_res.status_code, 200
new_id = create_res.json()
print(f"   Created Order ID: {new_id}")
print(f"   Created Order status: In Process")

# B. GET
get_res = requests.get(f"{BASE_URL}/orders/{new_id}")
assert get_res.status_code, 200
assert get_res.json()["status"], "In Process"

print("   Get Successfully after create new order.")

# C. UPDATE
payload["status"] = "Shipped"
put_res = requests.put(f"{BASE_URL}/orders/{new_id}", json=payload)
assert put_res.status_code, 200

# Verify status updated
get_res = requests.get(f"{BASE_URL}/orders/{new_id}")
assert get_res.status_code, 200
assert get_res.json()["status"], "Shipped"

print(f"   Updated successfully order status: {get_res.json()["status"]}")

# D. DELETE
del_res = requests.delete(f"{BASE_URL}/orders/{new_id}")
assert del_res.status_code, 200

# Verify delete successfully
get_res = requests.get(f"{BASE_URL}/orders/{new_id}")
assert get_res.status_code, 404

print("Delete successfully after a life circle.")
````

# Life circle test of orderdetails
## Create -> Get -> Update -> Delete
Test /orders/{orderNumber}/orderdetails	for GET, PUT, DELETE, and /orderdetails	for POST.

````python
# Setup: We need a temporary Order to hold the detail
order_payload = {
    "orderDate": "2025-05-05",
    "requiredDate": "2025-05-10",
    "status": "In Process",
    "customerNumber": EXISTING_CUSTOMER_ID
}
order_id = requests.post(f"{BASE_URL}/orders", json=order_payload).json()
product_id = EXISTING_PRODUCT_CODE

try:
    # A. CREATE DETAIL
    detail_payload = {
        "orderNumber": int(order_id),
        "productCode": product_id,
        "quantityOrdered": 50,
        "priceEach": 100.00,
        "orderLineNumber": 1
    }
    create_res = requests.post(f"{BASE_URL}/orderdetails", json=detail_payload)
    assert create_res.status_code, 200
    print("Created orderdetails with quantityOrdered: 50 and priceEach: 100")

    # B. GET SPECIFIC to verify create successfully
    url = f"{BASE_URL}/orders/{order_id}/orderdetails/{product_id}"
    get_res = requests.get(url)
    assert get_res.status_code, 200
    assert get_res.json()["quantityOrdered"], 50
    print("Created orderdetails verified successfully")

    # C. UPDATE
    detail_payload["quantityOrdered"] = 99
    detail_payload["priceEach"] = 110.00
    put_res = requests.put(url, json=detail_payload)
    assert put_res.status_code, 200
    print("Update to quantityOrdered: 99, priceEach: 110")

    # Verify update
    url = f"{BASE_URL}/orders/{order_id}/orderdetails/{product_id}"
    get_res = requests.get(url)
    assert get_res.status_code, 200
    assert get_res.json()["quantityOrdered"], 99
    assert get_res.json()["priceEach"], 110
    print("Updated orderdetails verified successfully")

    # D. DELETE
    del_res = requests.delete(url)
    assert del_res.status_code, 200
    
    # E. CONFIRM GONE
    check = requests.get(url)
    assert check.status_code, 404

    print("Delete successfully. Verified!")

finally:
    # Cleanup the temporary order
    requests.delete(f"{BASE_URL}/orders/{order_id}")
````

# 400 (Invalid input) error test of orderdetails

````python
payload = {
    "orderNumber": 99999999, 
    "productCode": EXISTING_PRODUCT_CODE,
    "quantityOrdered": 10,
    "priceEach": 10.00,
    "orderLineNumber": 1
}
# This should fail at DB level (Foreign Key Constraint, which is invalid input) -> 400
response = requests.post(f"{BASE_URL}/orderdetails", json=payload)
assert response.status_code, 400
````

# 422 (Input validation error) for customer

````python
payload = {
    "contactLastName": "Doe",
    "phone": "555-0000",
    "city": "Nowhere"
    # MISSING: customerName
}
response = requests.post(f"{BASE_URL}/customers", json=payload)

assert response.status_code, 422
print(f"\n[Customer] Missing Field Test: {response.status_code} OK")

"""
Expect 422: Sending a String for 'creditLimit' which expects a Float/Int.
"""
payload = {
    "customerName": "Bad Type Corp",
    "contactLastName": "Doe",
    "contactFirstName": "John",
    "phone": "555-1234",
    "addressLine1": "123 Street",
    "city": "City",
    "country": "USA",
    "creditLimit": "ONE MILLION DOLLARS"  # <--- INVALID TYPE
}
response = requests.post(f"{BASE_URL}/customers", json=payload)

assert response.status_code, 422
print(f"[Customer] Invalid Type Test: {response.status_code} OK")
````

# 400 error for orders

````python
"""
Expect 400: Creating an order for a customer ID that doesn't exist.
This triggers a Foreign Key Constraint failure in MySQL.
"""
payload = {
    "orderDate": "2025-01-01",
    "requiredDate": "2025-01-10",
    "status": "In Process",
    "customerNumber": 99999999  # <--- NON-EXISTENT CUSTOMER
}
response = requests.post(f"{BASE_URL}/orders", json=payload)

# This should be 400 (Bad Request)
assert response.status_code, 400
print(f"[Order] Invalid FK Test: Returned {response.status_code}")
````

# 400 error of updating(post) orderdetails

````python
"""
Expect 400: Adding a product that does not exist to an order.
"""
# First, create a temporary valid order to hold the detail
order_payload = {
    "orderDate": "2025-01-01", "requiredDate": "2025-01-10",
    "status": "In Process", "customerNumber": EXISTING_CUSTOMER_ID
}
order_id = requests.post(f"{BASE_URL}/orders", json=order_payload).json()

try:
    detail_payload = {
        "orderNumber": order_id,
        "productCode": "INVALID_CODE_999", # NON-EXISTENT PRODUCT
        "quantityOrdered": 10,
        "priceEach": 50.00,
        "orderLineNumber": 1
    }
    response = requests.post(f"{BASE_URL}/orderdetails", json=detail_payload)
    
    assert response.status_code, 400
    print(f"[OrderDetail] Invalid Product Test: Returned {response.status_code}. Success.")

finally:
    # Cleanup
    requests.delete(f"{BASE_URL}/orders/{order_id}")
````

# Add same orderdetail twice. Invalid update (400 error) for orderdetails

````python
"""
Expect 400: Adding the SAME product to the SAME order twice.
This violates the Composite Primary Key (orderNumber, productCode).
"""
# Setup: Create order
order_payload = {
    "orderDate": "2025-01-01", "requiredDate": "2025-01-10",
    "status": "In Process", "customerNumber": EXISTING_CUSTOMER_ID
}
order_id = requests.post(f"{BASE_URL}/orders", json=order_payload).json()

try:
    detail_payload = {
        "orderNumber": order_id,
        "productCode": EXISTING_PRODUCT_CODE,
        "quantityOrdered": 10,
        "priceEach": 50.00,
        "orderLineNumber": 1
    }
    
    # 1. First Insert (Should Succeed)
    res1 = requests.post(f"{BASE_URL}/orderdetails", json=detail_payload)
    assert res1.status_code, 200
    print(f"[OrderDetail] Initial create successfully")
    
    # 2. Second Insert (Should Fail - Duplicate Key)
    res2 = requests.post(f"{BASE_URL}/orderdetails", json=detail_payload)
    assert res2.status_code, 400
    print(f"[OrderDetail] Duplicate Entry Test: Returned {res2.status_code}. Success.")

finally:
    # Cleanup
    requests.delete(f"{BASE_URL}/orders/{order_id}")
````
